```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data_2d = pd.read_csv('./data/data_2d.csv')
data_mnist = pd.read_csv('./data/mnist.csv')
```

```python
def elbow_method(data, max_k=10):
    inertias = []
    for k in range(1, max_k+1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        inertias.append(kmeans.inertia_)
    
    plt.plot(range(1, max_k+1), inertias, marker='o')
    plt.xlabel('Liczba klastrów')
    plt.ylabel('Inertia')
    plt.title('Metoda łokciowa')
    plt.show()
```

```python
elbow_method(data_2d)
```

```python
optimal_k_2d = 3  # wartosc zalezy od liczby przelaman odczytanych z wykresu

kmeans_2d = KMeans(n_clusters=optimal_k_2d, random_state=42)
clusters_2d = kmeans_2d.fit_predict(data_2d)
```

```python
elbow_method(data_mnist)
```

```python
optimal_k_mnist = 5  # wartosc zalezy od liczby przelaman odczytanych z wykresu

kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=42)
clusters_mnist = kmeans_mnist.fit_predict(data_mnist)
```

```python
plt.scatter(data_2d.iloc[:, 0], data_2d.iloc[:, 1], c=clusters_2d, cmap='viridis')
plt.title('Klastrowanie danych 2D')
plt.show()
```

```python
pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(data_mnist)

plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=clusters_mnist, cmap='viridis')
plt.title('Klastrowanie danych MNIST (po redukcji PCA)')
plt.show()
```
